```python
import pandas as pd
import numpy as np
```

```python
trial_types = ['cue', 'match', 'mismatch']
```

```python
x = np.load('/Volumes/schnyer/Megan/adm_mem-fc/analysis/nbs/withhc/memmatch_fc_withhc_clean.npy', allow_pickle=True).flat[0]
```

```python
atlas_lut='data/power264_labels.tsv'

atlas = pd.read_csv(atlas_lut, sep='\t').set_index('index')
atlas = atlas.append({'regions': 'HC Sherman'}, ignore_index=True)
atlas.regions.unique()
```

```python
atlas = atlas.drop([3,9,183], axis=0)
atlas = atlas.reset_index(drop=True)
```

```python
hc = atlas.loc[atlas['regions'].str.contains('HC')].index.tolist()
dmn = atlas.loc[atlas['regions'].str.contains('Default')].index.tolist()
fpn = atlas.loc[atlas['regions'].str.contains('Fronto-parietal')].index.tolist()

dmn_hc = np.concatenate((dmn, hc))
fpn_hc = np.concatenate((fpn, hc))

dmn_fpn = np.concatenate((dmn, fpn))
```

```python
x['cue'].shape

x['cue'][dmn][:,dmn].shape

x['cue'][dmn][:,dmn].mean(axis=0).mean(axis=0).shape
```

```python
fcdf = pd.DataFrame()
fcdf['subject'] = x['subjects']

for trial_type in trial_types:
    
    x0 = x[trial_type][:,:,:-1]

    fcdf['dmnfc_%s' % trial_type] = np.nanmean(x0[dmn][:,dmn], axis=0).mean(axis=0)
    fcdf['fpnfc_%s' % trial_type] = np.nanmean(x0[fpn][:,fpn], axis=0).mean(axis=0)
    fcdf['dmnfpnfc_%s' % trial_type] = np.nanmean(x0[dmn][:,fpn], axis=0).mean(axis=0)
    fcdf['hcdmnfc_%s' % trial_type] = np.nanmean(x0[hc][:,dmn], axis=0).mean(axis=0)
    fcdf['hcfpnfc_%s' % trial_type] = np.nanmean(x0[hc][:,fpn], axis=0).mean(axis=0)

fcdf.head()
```

```python
y = np.load('/Volumes/schnyer/Megan/adm_mem-fc/analysis/nbs/withhc/rest_fc_withhc_clean.npy', allow_pickle=True).flat[0]
```

```python
atlas = pd.read_csv(atlas_lut, sep='\t').set_index('index')
atlas = atlas.append({'regions': 'HC Sherman'}, ignore_index=True)

atlas = atlas.drop([9], axis=0)
atlas = atlas.reset_index(drop=True)
```

```python
hc = atlas.loc[atlas['regions'].str.contains('HC')].index.tolist()
dmn = atlas.loc[atlas['regions'].str.contains('Default')].index.tolist()
fpn = atlas.loc[atlas['regions'].str.contains('Fronto-parietal')].index.tolist()

dmn_hc = np.concatenate((dmn, hc))
fpn_hc = np.concatenate((fpn, hc))

dmn_fpn = np.concatenate((dmn, fpn))
```

```python
fcdf_ = pd.DataFrame()
fcdf_['subject'] = y['subjects']

trial_type = 'rest'

y0 = y['rest']
fcdf_['dmnfc_%s' % trial_type] = np.nanmean(y0[dmn][:,dmn], axis=0).mean(axis=0)
fcdf_['fpnfc_%s' % trial_type] = np.nanmean(y0[fpn][:,fpn], axis=0).mean(axis=0)
fcdf_['dmnfpnfc_%s' % trial_type] = np.nanmean(y0[dmn][:,fpn], axis=0).mean(axis=0)
fcdf_['hcdmnfc_%s' % trial_type] = y0[hc][:,dmn].mean(axis=0).mean(axis=0)
fcdf_['hcfpnfc_%s' % trial_type] = y0[hc][:,fpn].mean(axis=0).mean(axis=0)

fcdf_.head()
```

```python
df = pd.merge(fcdf, fcdf_, on='subject', how='outer')
df.head()
```

```python
df.to_csv('data/fcdf.csv', index=None)
```
